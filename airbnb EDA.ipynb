**Task 4 - Data Cleaning**

```python
# Finding total no. of null values in each column
data.isnull().sum()
```

```python
# dropping all missing values
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
# Finding total no. of duplicates 
data.duplicated().sum()
```

```python
# Finding the duplicate rows 
data[data.duplicated()]
```

```python
# Dropping duplicate rows
data.drop_duplicates(inplace=True)
```

```python
data.duplicated().sum()
```

```python
# Checking the data types of columns
data.dtypes
```

```python
# Changing the data type of id and host_id  columns
data["id"] = data["id"].astype(object)
data["host_id"] = data["host_id"].astype(object)
```

```python
data.dtypes
```

**Task 5 - Data Analysis**

***5.1 - Univariate Analysis***

```python
data['price']
```

```python
# Price Distribution
# boxplot helps in identifying outliers
sns.boxplot(data=data, x="price")
plt.show()
```

```python
sns.histplot(data=data, x="price")
plt.show()
```

```python
# for handling the outlier observed in the boxplot
df = data[data["price"] < 1500]

# boxplot to view the new data distribution
sns.boxplot(data=df, x="price")
plt.show()
```

```python
# Histogram to view the total price distribution
sns.histplot(data=df, x='price', bins = 100)
plt.title("Price Distribution")
plt.ylabel("Frequency")
plt.figure(figsize=(8,5))
plt.show()
```

```python
df.dtypes
```

```python
# Histogram to view the Availability_365 distribution
sns.histplot(data=df, x='availability_365')
plt.title("Availability_365 Distribution")
plt.ylabel("Frequency")
plt.figure(figsize=(8,5))
plt.show()
```

```python
df.dtypes
```

```python
# Grouping neighbourhood_group column on the basis of average price
df.groupby(by='neighbourhood_group')['price'].mean()
```

**Feature Engineering**

```python
df.head()
```

```python
# adding new column 'Price per bed'
df['price per bed'] = df['price']/df['beds']
```

```python
df.head()
```

```python
# Grouping neighbourhood_group column on the basis of average price per bed
df.groupby(by='neighbourhood_group')['price per bed'].mean()
```

***5.2 - Bivariate Analysis***

```python
df.columns
```

```python
# Price distribution on the basis of neighbourhood_group
sns.barplot(data = df, x='neighbourhood_group', y='price', hue='room_type')
plt.show()
```

```python
# Price distribution on the basis of number_of_reviews
sns.scatterplot(data = df, x='number_of_reviews', y='price', hue='neighbourhood_group')
plt.title('Price Distribution v/s No. of Reviews')
plt.figure(figsize=(10,8))
plt.show()
```

```python
# Price Distribution based on room type
sns.barplot(data=df, x='room_type', y='price', hue='neighbourhood_group')
plt.title('Price Distribution based on Room Type')
plt.figure(figsize=(10,8))
plt.show()
```

**Advanced Analysis**

```python
sns.pairplot(data=df, vars=['price', 'minimum_nights', 'number_of_reviews', 'availability_365'], hue='room_type')
plt.show()
```

```python
#Geographical Distribution of AirBnb Listing
sns.scatterplot(data=df, x='latitude', y='longitude', hue='room_type')
plt.title("Geographical Distribution of AirBnb Listing")
plt.figure(figsize=(10, 8))
plt.show()
```

```python
# heat map - correlation of one variable with others for numerical column
corr = df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()
sns.heatmap(data=corr, annot=True)
plt.figure(figsize=(8, 6))
plt.show()
```
